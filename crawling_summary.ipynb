```python
import pandas as pd
```

```python
folder_path = 'C:/Users/tjdtn/inflearn-llm-application/big-contest/data/'

raw = pd.read_csv(folder_path+'jeju_store_keyword.csv')
crawling = raw.copy()
print(crawling.shape)
crawling.head(2)
```

```python
crawling['상위키워드'] = crawling['상위키워드'].str.strip('[|]').str.replace("'", "").str.split(', ')
```

```python
crawling.info()
```

```python
# 상위 키워드 컬럼의 데이터를 str로 만들기
for i in range(crawling.shape[0]):
    try:
        crawling.loc[i, '상위키워드'] = ', '.join(crawling.loc[i, '상위키워드'])
    except:
        crawling.loc[i, '상위키워드'] = None
```

```python
crawling.loc[0, '상위키워드']
```

```python
raw2 = pd.read_csv(folder_path+'new_주소별관광지.csv')
주변관광지 = raw2.copy()
print(주변관광지.shape)
주변관광지.head(2)
```

```python
# 리스트화
주변관광지['관광지명'] = 주변관광지['관광지명'].str.strip('[|]').str.replace("'", "").str.split(', ')
```

```python
# 주소 리스트
len(주변관광지['주소'].tolist())
```

```python
# 음식점에서 읍면동 반환하기
def check_address(address):
    for item in 주변관광지['주소'].tolist():
        if item in address:
            return item  # 포함된 읍면동을 반환
    return None  # 포함된 읍면동이 없으면 None 반환

crawling['작은주소'] = crawling['주소'].apply(check_address) # 73 (None 포함)
```

```python
crawling.head(2)
```

```python
# merge
spot_df = pd.merge(crawling, 주변관광지, left_on='작은주소', right_on='주소', how='outer')
print(spot_df.shape)
spot_df.head(2)
```

```python
# 필요없는 컬럼 삭제하기
spot_df.drop(columns=['작은주소', '주소_y'], inplace=True)
spot_df = spot_df.rename(columns={'주소_x':'주소'})
spot_df.head(2)
```

```python
# 관광지명을 주변관광지로 변경하기
spot_df = spot_df.rename(columns={'관광지명':'주변관광지'})

# 상위 키워드 컬럼의 데이터를 str로 만들기
for i in range(spot_df.shape[0]):
    try:
        spot_df.loc[i, '주변관광지'] = ', '.join(spot_df.loc[i, '주변관광지'])
    except:
        spot_df.loc[i, '주변관광지'] = None
        
spot_df.head(2)
```

```python
# 리뷰, 키워드 없애기
spot_df.drop(columns=['리뷰', '키워드'], inplace=True)
spot_df.head(2)
```

{까투리올레점}의 가게유형은 {일식튀김, 꼬치}, 주소는 {제주 서귀포시 동홍동 415-6번지 그레이스빌	}입니다.
주요 키워드는 {상위키워드}입니다. / 주요키워드에 대해서는 알 수 없습니다.(NaN일 경우)
주변 관광지는 {주변관광지} 입니다. / 주변 관광지에 대한 정보를 알 수 없습니다.
예약을 하고/하지 않고 방문한 사람들이 더 많고(더 많은 것의 비중) / 예약여부에 대해서 알 수 없고(NaN), (1일 경우)포장배달이 가능합니다./(0일 경우, NaN)포장배달에 대해서는 매장 문의가 필요합니다.
대부분 {가장 비중이 큰 대기시간 컬럼명}(비중), {두 번째}(비중) 대기하고 입장합니다. {}의 대기시간은 전체 매장 중 {순위}위이고, / 대기시간에 대한 정보는 매장에 문의가 필요합니다.(NaN)
대부분 {비중이 가장 큰 목적 컬럼명}(비중), {두번째}(비중) 목적으로 방문합니다. {비중이 가장 큰 목적 컬럼명} 목적은 전체 매장 중 {순위}위이고, {비중이 두 번째로 큰 목적 컬럼명} 목적은 전체 매장 중 {순위}위입니다./ 방문목적에 대해서는 알 수 없습니다.
{비중이 가장 큰 동반자 컬럼명}(비중)와 함께 가장 많이 방문하고, {두 번째}(비중)와 함께 두 번째로 많이 방문합니다. {비중이 가장 큰 동반자 컬럼명}동반자 유형은 전체 매장 중 {순위}이고, {두 번째} 동반자 유형은 전체 매장 중 {순위} 입니다.

```python
def make_summary(i):

    specific_row = spot_df.iloc[i]  # 인덱스 1 (2번째 행) 선택

    # 기본 정보 정리
    basic_info = f"{specific_row['가게명']}의 가게유형은 {specific_row['가게유형']}, 위치는 {specific_row['주소']} 입니다.\n"

    # 상위키워드 정리
    if specific_row['상위키워드'] == None:
        keyword_info = ''
    else:
        keyword_info = f"주요 키워드는 {specific_row['상위키워드']} 입니다.\n"

    # 주변관광지 정리
    if specific_row['주변관광지'] == None:
        around_info = ''
    else:
        around_info = f"주변 관광지는 {specific_row['주변관광지']} 입니다.\n"

    # 어떻게 먹었는냐에 대한 정리
    reservation_sorted_values = specific_row[['예약_무', '예약_유']].sort_values(ascending=False)
    
    if reservation_sorted_values.isna().all():
        reservation_info = "예약에 대한 정보는 매장 문의가 필요합니다."
    else:
        # 최대 비중의 예약유무 및 값
        max_reservation = reservation_sorted_values.idxmax()
        max_value_reservation = reservation_sorted_values.max()
        
        if max_value_reservation == 0.5:
            reservation_info = "예약을 하고, 하지 않고 방문한 사람들의 비율이 같고, "
        else:
            if max_reservation == '예약_무':
                reservation_info = f"예약을 하지 않고 방문한 사람들이 더 많고({max_value_reservation:.2f}), "
            elif max_reservation == '예약_유':
                reservation_info = f"예약을 하고 방문한 사람들이 더 많고({max_value_reservation:.2f}), "
        
    # 포장배달 유무
    if specific_row['포장배달_유무'] == 1:
        reservation_info = reservation_info + '포장배달이 가능합니다.\n'
        
    else:
        reservation_info = reservation_info + '포장배달에 대해서는 매장 문의가 필요합니다.\n'

    # 대기시간 컬럼 선택
    waiting_columns = ['바로 입장', '10분 이내', '30분 이내', '30분 이상', '1시간 이상', '2시간 이상']
    
    # 해당 행에서 비중 값을 정렬
    waiting_sorted_values = specific_row[waiting_columns].sort_values(ascending=False)
    
    if waiting_sorted_values.isna().all(): 
        waiting_info = "대기시간에 대한 정보는 매장 문의가 필요합니다." # null 값이면 정보 없음으로 표시
    else:
        # 최대 비중의 대기시간 및 값
        max_waiting = waiting_sorted_values.idxmax()
        max_value_waiting = waiting_sorted_values.max()
        
        # 두 번째 비중의 대기시간 및 값
        second_waiting = waiting_sorted_values.index[1]
        second_value_waiting = waiting_sorted_values.iloc[1]  # 두 번째 값
        
        if max_value_waiting == 1:
            waiting_info = f"대부분 {max_waiting}({max_value_waiting:.2f})의 대기시간을 가지고 입장합니다. "
            
            # 전체 컬럼의 값으로 정렬하여 순위 구하기
            max_waiting_ranked_values = spot_df[max_waiting].rank(ascending=False, method='dense')
            
            # 특정 행의 순위 가져오기
            max_waiting_rank = max_waiting_ranked_values.iloc[i]
            
            waiting_info = waiting_info + f"{max_waiting}의 대기시간을 가지고 방문한 손님의 비중은 전체 매장 중 {int(max_waiting_rank)}위 입니다.\n"
            
        else:
            waiting_info = f"대부분 {max_waiting}({max_value_waiting:.2f}), {second_waiting}({second_value_waiting:.2f})의 대기시간을 가지고 입장합니다. "

            # 전체 컬럼의 값으로 정렬하여 순위 구하기
            max_waiting_ranked_values = spot_df[max_waiting].rank(ascending=False, method='dense')
            second_waiting_ranked_values = spot_df[second_waiting].rank(ascending=False, method='dense')

            # 특정 행의 순위 가져오기
            max_waiting_rank = max_waiting_ranked_values.iloc[i]
            second_waiting_rank = second_waiting_ranked_values.iloc[i]
            
            waiting_info = waiting_info + f"{max_waiting}의 대기시간을 가지고 방문한 손님의 비중은 전체 매장 중 {int(max_waiting_rank)}위 이고, {second_waiting}의 대기시간을 가지고 방문한 손님의 비중은 전체 매장 중 {int(second_waiting_rank)}위 입니다.\n"

    # 목적 컬럼 선택
    purpose_columns = ['데이트', '친목', '나들이', '여행', '일상', '기념일', '회식', '가족모임', '비즈니스']

    # 해당 행에서 비중 값을 정렬
    purpose_sorted_values = specific_row[purpose_columns].sort_values(ascending=False)
        
    if purpose_sorted_values.isna().all():
        purpose_info = ""
    else:
        # 최대 비중의 목적 및 값
        max_purpose = purpose_sorted_values.idxmax()
        max_value_purpose = purpose_sorted_values.max()
        
        # 두 번째 비중의 시간 및 값
        second_purpose = purpose_sorted_values.index[1]  # 두 번째 인덱스
        second_value_purpose = purpose_sorted_values.iloc[1]  # 두 번째 값
        
        if max_value_purpose == 1:
            purpose_info = f"대부분 {max_purpose}({max_value_purpose:.2f}) 목적으로 방문합니다. "
            
            # 전체 컬럼의 값으로 정렬하여 순위 구하기
            max_purpose_ranked_values = spot_df[max_purpose].rank(ascending=False, method='dense')
            
            # 특정 행의 순위 가져오기
            max_purpose_rank = max_purpose_ranked_values.iloc[i]
            
            purpose_info = purpose_info + f"{max_purpose} 목적으로 방문한 손님의 비중은 전체 매장 중 {int(max_purpose_rank)}위 입니다.\n"
            
        else:
            purpose_info = f"대부분 {max_purpose}({max_value_purpose:.2f}), {second_purpose}({second_value_purpose:.2f}) 목적으로 방문합니다. "
        
            # 전체 컬럼의 값으로 정렬하여 순위 구하기
            max_purpose_ranked_values = spot_df[max_purpose].rank(ascending=False, method='dense')
            second_purpose_ranked_values = spot_df[second_purpose].rank(ascending=False, method='dense')

            # 특정 행의 순위 가져오기
            max_purpose_rank = max_purpose_ranked_values.iloc[i]
            second_purpose_rank = second_purpose_ranked_values.iloc[i]
            
            purpose_info = purpose_info + f"{max_purpose} 목적으로 방문한 손님의 비중은 전체 매장 중 {int(max_purpose_rank)}위 이고, {second_purpose} 목적으로 방문한 손님의 비중은 전체 매장 중 {int(second_purpose_rank)}위 입니다.\n"

    # 동반자 컬럼 선택
    with_columns = ['연인・배우자', '친구', '지인・동료', '혼자', '아이', '부모님', '반려동물', '친척・형제', '기타']

    # 해당 행에서 비중 값을 정렬
    with_sorted_values = specific_row[with_columns].sort_values(ascending=False)

    if with_sorted_values.isna().all():
        with_info = ""
    else:
        # 최대 비중의 동반자 및 값
        max_with = with_sorted_values.idxmax()
        max_value_with = with_sorted_values.max()
        
        # 두 번째 비중의 동반자 및 값
        second_with = with_sorted_values.index[1]
        second_value_with = with_sorted_values.iloc[1]
        
        if max_value_with == 1:
            with_info = f"{max_with}({max_value_with:.2f})와 함께 가장 많이 방문합니다. "
            
            # 전체 컬럼의 값으로 정렬하여 순위 구하기
            max_with_ranked_values = spot_df[max_with].rank(ascending=False, method='dense')
            
            # 특정 행의 순위 가져오기
            max_with_rank = max_with_ranked_values.iloc[i]
            
            with_info = with_info + f"{max_with} 동반자 유형은 전체 매장 중 {int(max_with_rank)}위 입니다."
            
        else:   
            with_info = f"{max_with}({max_value_with:.2f})와 함께 가장 많이 방문하고, {second_with}({second_value_with:.2f})와 함께 두 번째로 많이 방문합니다. "
            
            # 전체 컬럼의 값으로 정렬하여 순위 구하기
            max_with_ranked_values = spot_df[max_with].rank(ascending=False, method='dense')
            second_with_ranked_values = spot_df[second_with].rank(ascending=False, method='dense')

            # 특정 행의 순위 가져오기
            max_with_rank = max_with_ranked_values.iloc[i]
            second_with_rank = second_with_ranked_values.iloc[i]
            
            with_info = with_info + f"{max_with} 동반자 유형은 전체 매장 중 {int(max_with_rank)}위 이고, {second_with} 동반자 유형은 전체 매장 중 {int(second_with_rank)}위 입니다."

    summary_info = basic_info + keyword_info + around_info + reservation_info + waiting_info + purpose_info + with_info
    return summary_info
```
