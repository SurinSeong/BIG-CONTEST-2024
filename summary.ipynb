```python
original_12.columns = ['기준연월', '가맹점명', '개설일자', '업종', '주소', '이용건수구간', '이용금액구간', '건당평균이용금액구간',
       '월_이용건수비중', '화_이용건수비중', '수_이용건수비중', '목_이용건수비중', '금_이용건수비중', '토_이용건수비중',
       '일_이용건수비중', '05-11시_이용건수비중', '12-13시_이용건수비중', '14-17시_이용건수비중',
       '18-22시_이용건수비중', '23-04시_이용건수비중', '현지인_이용건수비중', '남성_이용건수비중',
       '여성_이용건수비중', '10-20대_회원비중', '30대_회원비중', '40대_회원비중', '50대_회원비중',
       '60대이상_회원비중']
```

```python
# 순위 구하기
original_12['남성_이용건수순위'] = original_12.남성_이용건수비중.rank(method='dense', ascending=False).astype(int)
original_12['여성_이용건수순위'] = original_12.여성_이용건수비중.rank(method='dense', ascending=False).astype(int)

original_12['10-20대_회원순위'] = original_12['10-20대_회원비중'].rank(method='dense', ascending=False).astype(int)
original_12['30대_회원순위'] = original_12['30대_회원비중'].rank(method='dense', ascending=False).astype(int)
original_12['40대_회원순위'] = original_12['40대_회원비중'].rank(method='dense', ascending=False).astype(int)
original_12['50대_회원순위'] = original_12['50대_회원비중'].rank(method='dense', ascending=False).astype(int)
original_12['60대이상_회원순위'] = original_12['60대이상_회원비중'].rank(method='dense', ascending=False).astype(int)
```

```python
# 이용건수가 가장 많은 요일
# 각 행별로 가장 큰 비중의 요일 찾기
original_12['최대_이용요일'] = original_12[['월_이용건수비중', '화_이용건수비중', '수_이용건수비중', '목_이용건수비중', 
                        '금_이용건수비중', '토_이용건수비중', '일_이용건수비중']].idxmax(axis=1).str.replace('_이용건수비중', '')

original_12['최대_이용시간'] = original_12[['05-11시_이용건수비중', '12-13시_이용건수비중', '14-17시_이용건수비중',
       '18-22시_이용건수비중', '23-04시_이용건수비중']].idxmax(axis=1).str.replace('_이용건수비중', '')

original_12['최대_이용성별'] = original_12[['남성_이용건수비중', '여성_이용건수비중',]].idxmax(axis=1).str.replace('_이용건수비중', '')


original_12['최대_이용나이'] = original_12[['10-20대_회원비중', '30대_회원비중', '40대_회원비중', '50대_회원비중', '60대이상_회원비중']].idxmax(axis=1).str.replace('_회원비중', '')

# 최대 비중을 가진 나이대의 값을 NaN으로 변경하여 두 번째로 큰 비중 찾기
without_max = original_12[['10-20대_회원비중', '30대_회원비중', '40대_회원비중', '50대_회원비중', '60대이상_회원비중']].mask(
    original_12[['10-20대_회원비중', '30대_회원비중', '40대_회원비중', '50대_회원비중', '60대이상_회원비중']].eq(original_12[['10-20대_회원비중', '30대_회원비중', '40대_회원비중', '50대_회원비중',
       '60대이상_회원비중']].max(axis=1), axis=0))

# 두 번째로 큰 비중의 나이대 찾기
original_12['두번째_이용나이'] = without_max.idxmax(axis=1).str.replace('_회원비중', '')
```

```python
def make_summary(i):
    # 특정 행(예: 2번째 행, 인덱스 1)에서 최대 및 두 번째 비중의 요일과 그 값을 구하기
    specific_row = original_12.iloc[i]  # 인덱스 1 (2번째 행) 선택

    basic_info = f"{specific_row['가맹점명']}의 업종은 {specific_row['업종']}, 위치는 {specific_row['주소']} 입니다."
    used_info = f"이용건수는 동일 업종 내 상위 {specific_row['이용건수구간'][2:]}이고, 현지인 이용건수 비중은 {specific_row['현지인_이용건수비중']:.2f}입니다."
    gender_info = f"{specific_row['최대_이용성별']}의 이용건수가 더 높습니다. 남성 이용건수는 전체 가맹점 중 {specific_row['남성_이용건수순위']}위({specific_row['남성_이용건수비중']:.2f})이고, 여성 이용건수는 전체 가맹점 중 {specific_row['여성_이용건수순위']}위({specific_row['여성_이용건수비중']:.2f})입니다."

    # 요일별 비중 컬럼만 선택 (특정 컬럼들만 비교)
    weekday_columns = ['월_이용건수비중', '화_이용건수비중', '수_이용건수비중', '목_이용건수비중',
                    '금_이용건수비중', '토_이용건수비중', '일_이용건수비중']

    time_columns = ['05-11시_이용건수비중', '12-13시_이용건수비중', '14-17시_이용건수비중',
                    '18-22시_이용건수비중', '23-04시_이용건수비중']

    age_columns = ['10-20대_회원비중', '30대_회원비중', '40대_회원비중', '50대_회원비중', '60대이상_회원비중']

    # 해당 행에서 비중 값을 정렬하여 두 번째로 큰 값을 찾기
    weekday_sorted_values = specific_row[weekday_columns].sort_values(ascending=False)
    time_sorted_values = specific_row[time_columns].sort_values(ascending=False)
    age_sorted_values = specific_row[age_columns].sort_values(ascending=False)

    # 최대 비중의 요일 및 값
    max_weekday = weekday_sorted_values.idxmax().replace('_이용건수비중', '')
    max_value_weekday = weekday_sorted_values.max()

    # 최대 비중의 시간 및 값
    max_time = time_sorted_values.idxmax().replace('_이용건수비중', '')
    max_value_time = time_sorted_values.max()

    # 최대 비중의 나이대 및 값
    max_age = age_sorted_values.idxmax().replace('_회원비중', '')
    max_value_age = age_sorted_values.max()

    # 두 번째 비중의 요일 및 값
    second_max_weekday = weekday_sorted_values.index[1].replace('_이용건수비중', '')  # 두 번째 인덱스
    second_max_value_weekday = weekday_sorted_values.iloc[1]  # 두 번째 값

    # 두 번째 비중의 시간 및 값
    second_max_time = time_sorted_values.index[1].replace('_이용건수비중', '')  # 두 번째 인덱스
    second_max_value_time = time_sorted_values.iloc[1]  # 두 번째 값

    # 두 번째 비중의 나이대 및 값
    second_max_age = age_sorted_values.index[1].replace('_회원비중', '')  # 두 번째 인덱스
    second_max_value_age = age_sorted_values.iloc[1]  # 두 번째 값

    # 특정 행(예: 2번째 행, 인덱스 1)에서 '화_이용건수비중' 컬럼의 순위 구하기
    max_age_column_name = f"{max_age}_회원비중"  # 특정 컬럼 이름
    second_age_column_name = f"{second_max_age}_회원비중"

    # 전체 컬럼의 값으로 정렬하여 순위 구하기
    max_age_ranked_values = original_12[max_age_column_name].rank(ascending=False, method='dense').astype(int)
    second_age_ranked_values = original_12[second_age_column_name].rank(ascending=False, method='dense').astype(int)

    # 특정 행의 순위 가져오기
    max_age_rank = max_age_ranked_values.iloc[i]
    second_age_rank = second_age_ranked_values.iloc[i]

    weekday_info = f"이용건수가 가장 많은 요일은 {max_weekday}요일({max_value_weekday:.2f}), 두 번째로 많은 요일은 {second_max_weekday}요일({second_max_value_weekday:.2f})입니다."
    time_info = f"이용건수가 가장 많은 시간대는 {max_time}({max_value_time:.2f}), 두 번째로 많은 시간대는 {second_max_time}({second_max_value_time:.2f})입니다."
    age_info = f"회원이 가장 많은 나이대는 {max_age}({max_value_age:.2f})이고 전체 가맹점 중 {max_age_rank}위입니다. 회원이 두 번째로 많은 나이대는 {second_max_age}({second_max_value_age:.2f})이고 전체 가맹점 중 {second_age_rank}위입니다."

    summary_info = basic_info + "\n" + used_info + "\n" + weekday_info + "\n" + time_info + "\n" + gender_info + "\n" + age_info

    return summary_info
```

```python
original_12['요약'] = None

for i in range(original_12.shape[0]):
    original_12.iloc[i, -1] = make_summary(i)
```
