```python
import pandas as pd
```

```python
folder_path = 'C:/Users/tjdtn/inflearn-llm-application/big-contest/data/'

raw = pd.read_csv(folder_path+'서귀포시_관광지주소.csv')
seoguipo = raw.copy()
print(seoguipo.shape)
seoguipo.head(2)
```

```python
raw2 = pd.read_csv(folder_path+'제주특별자치도 서귀포시_도로명주소고시현황_20240401.csv', encoding='cp949')
seoguipo_all_address = raw2.copy()
print(seoguipo_all_address.shape)
seoguipo_all_address.head(2)
```

```python
# 필요한 컬럼만 남기기
seoguipo_all_address = seoguipo_all_address[['종전주소', '도로명주소']]
```

```python
seoguipo_all_address['종전주소'] = seoguipo_all_address['종전주소'].str.replace('제주특별자치도 ', '')
seoguipo_all_address['도로명주소'] = seoguipo_all_address['도로명주소'].str.replace('제주특별자치도 ', '')
seoguipo_all_address.head(2)
```

```python
# merge 시도
merge_df = pd.merge(seoguipo, seoguipo_all_address, left_on='관광지주소', right_on='도로명주소', how='left')
print(merge_df.shape)
merge_df.head(2)
```

```python
# 안나온거 확인하기
merge_df[merge_df['종전주소'].isnull() == True]
```

```python
# 나온거 따로 저장 준비
seoguipo_notnull = merge_df[merge_df['종전주소'].isnull() != True][['관광지명', '종전주소']].reset_index(drop=True)
```

```python
seoguipo_notnull['종전주소'] = seoguipo_notnull['종전주소'].str.split('， ')
seoguipo_notnull['주소개수'] = seoguipo_notnull['종전주소'].apply(len)

seoguipo_notnull[seoguipo_notnull['주소개수'] != 1]
```

```python
def select_first(address_list):
    address = address_list[0]
    return address

seoguipo_notnull['종전주소'] = seoguipo_notnull['종전주소'].apply(select_first)
seoguipo_notnull = seoguipo_notnull.rename(columns={'종전주소': '관광지주소'})
seoguipo_notnull.head(2)
```

```python
seoguipo_notnull = seoguipo_notnull[['관광지명', '관광지주소']]

# 저장하기
seoguipo_notnull.to_csv(folder_path+'new_서귀포시_관광지주소.csv', index=False)
```

```python
merge_df[merge_df['종전주소'].isnull() == True][['관광지명', '관광지주소']].reset_index(drop=True).to_csv(folder_path+'찾아야할서귀포시_관광지주소.csv', index=False)
```

## 모든 제주시, 서귀포시 주소 합치기

```python
raw3 = pd.read_csv(folder_path+'찾아야할서귀포시_관광지주소.csv')
seoguipo_notnull2 = raw3.copy()
print(seoguipo_notnull2.shape)
seoguipo_notnull2.head(2)
```

```python
# new_~ 합치기
all_seoguipo = pd.concat([seoguipo_notnull, seoguipo_notnull2]).reset_index(drop=True)
```

```python
# 제주시 주소
raw4 = pd.read_csv(folder_path+'제주시_관광지주소.csv')
all_jeju = raw4.copy()
print(all_jeju.shape)
all_jeju.head(2)
```

```python
# 필요한 컬럼만 남기기
all_jeju = all_jeju[['관광지명', '지번주소']]

# 컬럼명 변경하기
all_jeju = all_jeju.rename(columns={'지번주소':'관광지주소'})
all_jeju.head(2)
```

```python
# 다시 모든것 합치기
all_spot = pd.concat([all_seoguipo, all_jeju]).reset_index(drop=True)
all_spot.head(2)
```

```python
# 일단 저장
all_spot.to_csv(folder_path+'new_제주도관광지주소.csv', index=False)
```
