# 연습용 db 구축해보기

```python
from langchain_core.documents import Document
```

```python
# Document 객체 리스트 생성하기
documents = []

for item in oliveyoung_dict:
    # 'Markdown' 필드의 텍스트 가져오기
    review_markdown = item.get('Markdown')
    # 'review_summary'가 str인지 확인, 그렇지 않으면 str로 변환하는 작업
    if not isinstance(review_markdown, str):
        review_markdown = str(review_markdown)

    # 메타데이터 생성하기
    metadata = {
        'name':item.get('name'),
        'brand':item.get('brand'),
        'price':item.get('price'),
        'link':item.get('url'),
        'summary':item.get('Summary'),
    }

    # Document 객체 생성
    document = Document(page_content=review_markdown, metadata=metadata)
    documents.append(document)
```

## 생성된 documents 리스트를 사용하여 Chroma DB에 저장하는 작업을 진행합니다.

#### chunking

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
# 텍스트 chunking splitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)

# chunking 하기
split_documents = text_splitter.split_documents(documents)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
# 허깅페이스모델을 사용해서 임베딩
embedding_model = HuggingFaceEmbeddings(model_name="upskyy/bge-m3-Korean")
```

```python
from langchain_chroma import Chroma
```

```python
# 분할된 문서를 Chroma에 저장
def save_to_chroma(doc_chunks, collection_name, persist_directory):
    chunked_documents = []
    for chunk in doc_chunks:
        chunked_documents.append(chunk)
        if len(chunked_documents) >= 100:
            try:
                Chroma.from_documents(
                    documents=chunked_documents,
                    embedding=embedding_model,
                    collection_name=collection_name,
                    persist_directory=persist_directory,
                )
            except Exception as e:
                print(f"Error saving documents to Chroma: {e}")
            finally:
                chunked_documents = []  # 저장 후 리스트 초기화

    # 마지막 남은 문서들 저장
    if chunked_documents:
        try:
            Chroma.from_documents(
                documents=chunked_documents,
                embedding=embedding_model,
                collection_name=collection_name,
                persist_directory=persist_directory,
            )
        except Exception as e:
            print(f"Error saving remaining documents to Chroma: {e}")


# Chroma에 저장
save_to_chroma(
    split_documents,
    collection_name="product_list",  # 적절한 이름으로 변경
    persist_directory=folder_path+"product_list_prac",  # 적절한 경로로 변경
)
```
