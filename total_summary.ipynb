## mct 요약 먼저 수정하기
* 모든 것에 순위 매기기
    * 현지인 이용건수 순위
    * 이용건수가 많은 요일 순위
    * 시간대 순위

* null값이 없음. 따라서, 'None이면'이라는 조건을 달지 않고 만들 수 있음.

```python
df['이용건수구간'].unique()
```

```python
# 상위 라는 말을 붙이기 위해 '1_상위 10% 이하'의 '상위' 삭제
df['이용건수구간'] = df['이용건수구간'].str.replace('상위 ', '')
df['이용건수구간'].unique()
```

```python
df.loc[0, 'keyword_요약'].split('\n')
```

```python
def total_summary(i):    
    specific_row = df.iloc[i]

    # 기본 정보는 mct_요약에서 가져오기
    basic_info = specific_row['mct_요약'].split('\n')[0] + '\n'
    
    # 상위 키워드
    keyword_info = specific_row['keyword_요약'].split('\n')[1] + '\n'
    
    # 주변관광지
    spot_info = specific_row['keyword_요약'].split('\n')[2] + '\n'
    
    ## 이용건수
    used_info = f"이용건수는 동일 업종 내 상위 {specific_row['이용건수구간'][2:]}이고, "

    ## 현지인_이용건수비중
    # 전체 컬럼의 값으로 정렬하여 순위 구하기
    local_ranked_values = df['현지인_이용건수비중'].rank(ascending=False, method='dense')
    used_info = used_info + f"현지인 이용건수 비중은 {specific_row['현지인_이용건수비중']:.2f}로, 순위는 전체 가게 중 {int(local_ranked_values.iloc[i])}위 입니다.\n"

    ## 요일별 비중 컬럼만 선택 (특정 컬럼들만 비교)
    weekday_columns = ['월_이용건수비중', '화_이용건수비중', '수_이용건수비중', '목_이용건수비중', '금_이용건수비중', '토_이용건수비중', '일_이용건수비중']

    # 해당 행에서 비중 값을 정렬
    weekday_sorted_values = specific_row[weekday_columns].sort_values(ascending=False)

    # 최대 비중의 요일 및 값
    max_weekday = weekday_sorted_values.idxmax()
    max_value_weekday = weekday_sorted_values.max()

    # 전체 컬럼의 값으로 정렬하여 순위 구하기
    max_weekday_ranked_values = df[max_weekday].rank(ascending=False, method='dense')

    weekday_info = f"이용건수가 가장 많은 요일은 {max_weekday[0]}요일({max_value_weekday:.2f}), 전체 가게 중 {int(max_weekday_ranked_values.iloc[i])}위 입니다."

    if max_value_weekday == 1:
        weekday_info = weekday_info + '\n'
    else:
        # 두 번째 비중의 요일 및 값
        second_weekday = weekday_sorted_values.index[1]  # 두 번째 인덱스
        second_value_weekday = weekday_sorted_values.iloc[1]  # 두 번째 값
        
        # 전체 컬럼의 값으로 정렬하여 순위 구하기
        second_weekday_ranked_values = df[second_weekday].rank(ascending=False, method='dense')
        
        weekday_info = weekday_info + f" 두 번째로 많은 요일은 {second_weekday[0]}요일({second_value_weekday:.2f}), 전체 가게 중 {int(second_weekday_ranked_values.iloc[i])}위 입니다.\n"

    ## 시간대
    time_columns = ['05-11시_이용건수비중', '12-13시_이용건수비중', '14-17시_이용건수비중', '18-22시_이용건수비중', '23-04시_이용건수비중']

    # 해당 행에서 비중 값을 정렬
    time_sorted_values = specific_row[time_columns].sort_values(ascending=False)

    # 최대 비중의 시간대 및 값
    max_time = time_sorted_values.idxmax()
    max_value_time = time_sorted_values.max()

    # 전체 컬럼의 값으로 정렬하여 순위 구하기
    max_time_ranked_values = df[max_time].rank(ascending=False, method='dense')

    time_info = f"이용건수가 가장 많은 시간대는 {max_time.split('_')[0]}({max_value_time:.2f}), 전체 가게 중 {int(max_time_ranked_values.iloc[i])}위 입니다."

    if max_value_time == 1:
        time_info = time_info + '\n'
    else:
        # 두 번째 비중의 시간대 및 값
        second_time = time_sorted_values.index[1]  # 두 번째 인덱스
        second_value_time = time_sorted_values.iloc[1]  # 두 번째 값
        
        # 전체 컬럼의 값으로 정렬하여 순위 구하기
        second_time_ranked_values = df[second_time].rank(ascending=False, method='dense')
        
        time_info = time_info + f" 두 번째로 많은 시간대는 {second_time.split('_')[0]}({second_value_time:.2f}), 전체 가게 중 {int(second_time_ranked_values.iloc[0])}위 입니다.\n"

    ## 성별
    gender_columns = ['남성_이용건수비중', '여성_이용건수비중']

    # 해당 행에서 비중 값을 정렬
    gender_sorted_values = specific_row[gender_columns].sort_values(ascending=False)

    # 최대 비중의 성별 및 값
    max_gender = gender_sorted_values.idxmax()
    max_value_gender = gender_sorted_values.max()

    # 두 번째 비중의 시간대 및 값
    second_gender = gender_sorted_values.index[1]  # 두 번째 인덱스
    second_value_gender = gender_sorted_values.iloc[1]  # 두 번째 값

    # 전체 컬럼의 값으로 정렬하여 순위 구하기
    max_gender_ranked_values = df[max_gender].rank(ascending=False, method='dense')
    second_gender_ranked_values = df[second_gender].rank(ascending=False, method='dense')

    gender_info = f"{max_gender.split('_')[0]}의 이용건수가 더 높습니다. {max_gender.split('_')[0]} 이용건수는 전체 가맹점 중 {int(max_gender_ranked_values.iloc[i])}위({max_value_gender:.2f})이고, {second_gender.split('_')[0]} 이용건수는 전체 가맹점 중 {int(second_gender_ranked_values.iloc[i])}위({second_value_gender:.2f})입니다.\n"

    ## 나이대
    age_columns = ['10-20대_회원비중', '30대_회원비중', '40대_회원비중', '50대_회원비중', '60대이상_회원비중']

    # 해당 행에서 비중 값을 정렬
    age_sorted_values = specific_row[age_columns].sort_values(ascending=False)

    # 최대 비중의 나이대 및 값
    max_age = age_sorted_values.idxmax()
    max_value_age = age_sorted_values.max()

    # 전체 컬럼의 값으로 정렬하여 순위 구하기
    max_age_ranked_values = df[max_age].rank(ascending=False, method='dense')

    age_info = f"회원이 가장 많은 나이대는 {max_age.split('_')[0]}({max_value_age:.2f}), 전체 가게 중 {int(max_age_ranked_values.iloc[i])}위 입니다."

    if max_value_age == 1:
        age_info = age_info + '\n'
    else:
        # 두 번째 비중의 나이대 및 값
        second_age = age_sorted_values.index[1]  # 두 번째 인덱스
        second_value_age = age_sorted_values.iloc[1]  # 두 번째 값
        
        # 전체 컬럼의 값으로 정렬하여 순위 구하기
        second_age_ranked_values = df[second_age].rank(ascending=False, method='dense')
        
        age_info = age_info + f" 두 번째로 많은 나이대는 {second_age.split('_')[0]}({second_value_age:.2f}), 전체 가게 중 {int(second_age_ranked_values.iloc[i])}위 입니다.\n"
        
#-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------
    remain_info = '\n'.join(specific_row['keyword_요약'].split('\n')[3:])

    all_summary = basic_info + keyword_info + spot_info + used_info + weekday_info + time_info + gender_info + age_info + remain_info

    return all_summary

print(total_summary(0))
```

```python
df['요약'] = None

for i in range(df.shape[0]):
    df.iloc[i, -1] = total_summary(i)
    
df.head(2)
```

```python
df.to_csv(folder_path+'JEJU_MCT_KEYWORD_v2.csv', index=False)
```
